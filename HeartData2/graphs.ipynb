```python
import numpy as np
import matplotlib.pyplot as plt

# set width of bar
barWidth = 0.25
fig = plt.subplots(figsize =(12, 8))

# set height of bar
OG = [0.88, 0.11, 0.5, 0.06, 0.79, 0.11]
Fake = [0.88,0.15,0.58,0.08,0.80,0.11]
Dfake = [0.87,0.17,0.39,0.11,0.78,0.12]

# Set position of bar on X axis
br1 = np.arange(len(OG))
br2 = [x + barWidth for x in br1]
br3 = [x + barWidth for x in br2]

# Make the plot
plt.bar(br1, OG, color ='r', width = barWidth,
		edgecolor ='grey', label ='Original Data')
plt.bar(br2, Fake, color ='g', width = barWidth,
		edgecolor ='grey', label ='Fake Data')
plt.bar(br3, Dfake, color ='b', width = barWidth,
		edgecolor ='grey', label ='Fake Data Of positive Cases')

# Adding Xticks
plt.xlabel('KNeighbor Classifier', fontweight ='bold', fontsize = 15)
plt.ylabel('White', fontweight ='bold', fontsize = 15)
plt.xticks([r + barWidth for r in range(len(OG))],
		['Accuracy', 'F1 Score', 'Precision', 'Recall', 'Rand Score', 'Mean Abs Error'])

plt.legend()
plt.show()
```
