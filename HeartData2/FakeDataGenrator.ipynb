```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import f1_score
from sklearn.model_selection import train_test_split
import warnings
from numpy.random import randn
warnings.filterwarnings('ignore')
import tensorflow as tf
import sklearn.neighbors
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from keras.models import Sequential
from sklearn import metrics
from tensorflow import keras
from keras.layers import Dense, Input
from keras.models import Model
from keras.optimizers import Adam
```

```python
from keras.models import load_model
model =load_model(r'C:\Users\parab\HeartData2\trained_generated_blackdisease_model.h5')
```

```python
def generate_latent_points(latent_dim, n_samples):
    x_input = randn(latent_dim * n_samples)
    x_input = x_input.reshape(n_samples, latent_dim)    
    return x_input
```

```python
latent_points = generate_latent_points(10, 400)
X = model.predict(latent_points)
data_fake = pd.DataFrame(data=X,  columns=['HeartDisease', 'BMI', 'Smoking', 'AlcoholDrinking', 'Stroke', 'PhysicalHealth', 'MentalHealth', 'DiffWalking', 'Sex', 'AgeCategory', 'Diabetic', 'PhysicalActivity','GenHealth','SleepTime','Asthma','KidneyDisease','SkinCancer'] )
data_fake.head()
```

```python
outcome_mean = data_fake.HeartDisease.mean()
data_fake['HeartDisease'] = True
data_fake["HeartDisease"] = data_fake["HeartDisease"].astype(int)
outcome_mean = data_fake.Smoking.mean()
data_fake['Smoking'] = data_fake['Smoking'] > outcome_mean
data_fake["Smoking"] = data_fake["Smoking"].astype(int)
outcome_mean = data_fake.AlcoholDrinking.mean()
data_fake['AlcoholDrinking'] = data_fake['AlcoholDrinking'] > outcome_mean
data_fake["AlcoholDrinking"] = data_fake["AlcoholDrinking"].astype(int)
outcome_mean = data_fake.Stroke.mean()
data_fake['Stroke'] = data_fake['Stroke'] > outcome_mean
data_fake["Stroke"] = data_fake["Stroke"].astype(int)
outcome_mean = data_fake.DiffWalking.mean()
data_fake['DiffWalking'] = data_fake['DiffWalking'] > outcome_mean
data_fake["DiffWalking"] = data_fake["DiffWalking"].astype(int)
outcome_mean = data_fake.Sex.mean()
data_fake['Sex'] = data_fake['Sex'] > outcome_mean
data_fake["Sex"] = data_fake["Sex"].astype(int)
outcome_mean = data_fake.Diabetic.mean()
data_fake['Diabetic'] = data_fake['Diabetic'] > outcome_mean
data_fake["Diabetic"] = data_fake["Diabetic"].astype(int)
outcome_mean = data_fake.PhysicalActivity.mean()
data_fake['PhysicalActivity'] = data_fake['PhysicalActivity'] > outcome_mean
data_fake["PhysicalActivity"] = data_fake["PhysicalActivity"].astype(int)
outcome_mean = data_fake.Asthma.mean()
data_fake['Asthma'] = data_fake['Asthma'] > outcome_mean
data_fake["Asthma"] = data_fake["Asthma"].astype(int)
outcome_mean = data_fake.KidneyDisease.mean()
data_fake['KidneyDisease'] = data_fake['KidneyDisease'] > outcome_mean
data_fake["KidneyDisease"] = data_fake["KidneyDisease"].astype(int)
outcome_mean = data_fake.SkinCancer.mean()
data_fake['SkinCancer'] = data_fake['SkinCancer'] > outcome_mean
data_fake["SkinCancer"] = data_fake["SkinCancer"].astype(int)
```

```python
outcome_mean = data_fake.GenHealth.mean()
outcome_std=data_fake.GenHealth.std()
for i in range(len(data_fake['GenHealth'])):
    if data_fake['GenHealth'][i] > outcome_mean+2*outcome_std:
        data_fake['GenHealth'][i]=5
    elif data_fake['GenHealth'][i] > outcome_mean+outcome_std:
        data_fake['GenHealth'][i]=4
    elif data_fake['GenHealth'][i] < outcome_mean-2*outcome_std:
        data_fake['GenHealth'][i]=1
    elif data_fake['GenHealth'][i] < outcome_mean-outcome_std:
        data_fake['GenHealth'][i]=2
    else:
        data_fake['GenHealth'][i]=3
```
